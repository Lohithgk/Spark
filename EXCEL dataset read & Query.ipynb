To work with Excel, Install following dependency from the library.

1. click clusters, then double click the cluster you want to work with.
2. click Libraries, click Install New
3. click Maven, In Coordinates and Install
4. com.crealytics:spark-excel_2.12:3.1.2_0.17.0

```
df1 =spark.read\
.format("com.crealytics.spark.excel")\
.option("header", True)\
.option("inferSchema",True)\
.option("path", "dbfs:/FileStore/tables/Person_Person_excel.xlsx")\
.load()
```

```
# Create a view or table
temp_table_name = "Person_Person_excel"

df1.createOrReplaceTempView(temp_table_name)
```

```
%sql
/* Query the created temp table in a SQL cell */
select * from `Person_Person_excel`
```

```
spark.sql("""
select * from Person_Person_excel
""").show()
```
