```
# File location and type
file_location = "dbfs:/FileStore/tables/Person_Person.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","
```

```
# The applied options are for CSV files. For other file types, these will be ignored.
df = spark\
    .read\
    .format(file_type)\
    .option("inferSchema", infer_schema)\
    .option("header", first_row_is_header)\
    .option("sep", delimiter)\
    .load(file_location)
```

```
# In PySpark, select() function is used to select single, multiple by passing the column names you wanted to select
# select() is a transformation function hence it returns a new DataFrame with the selected columns.
# dropDuplicates() function which returns a new DataFrame after removing duplicate rows.

df1 = df.select("EmailPromotion").dropDuplicates(["EmailPromotion"]).show()
```

```
# Drop selected coloumns from the dataframe using .drop()

df2 = df.drop("BusinessEntityID", "PersonType", "NameStyle", "Title", "FirstName")
```
