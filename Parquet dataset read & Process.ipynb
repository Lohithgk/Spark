***Apache Parquet file Read and Processing***

Apache Parquet is a columnar file format that provides optimizations to speed up queries and is a far more efficient file format than CSV or JSON

Spark SQL provides support for both reading and writing Parquet files that automatically capture the schema of the original data.

It also reduces data storage by 75% on average,
Reduces IO operations,
Fetches specific columns that you need to access,
It consumes less space.

```
# File location and type
file_location = "dbfs:/FileStore/tables/userdata.parquet"
file_type = "parquet"

# Read Parquet file using Dataframe DataReader API
df = spark\
    .read\
    .parquet(file_location)

display(df)
```

```
# Create a temp table
df.createOrReplaceTempView('userdata_Temptable')
```

```
%sql
/* Query the created temp table in a SQL cell */
select
concat(first_name,' ',last_name) as Full_Name,
email,
ip_address,
gender
FROM userdata_Temptable 
WHERE gender = 'Female'
```
