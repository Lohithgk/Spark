Write Spark DataFrame to CSV file

compression - Use the compression codec option when you want to compress a CSV file while writing to disk to reduce disk space. It supports the following values none, bzip2, gzip, lz4, snappy and deflate, and Default’s to null.

encoding - specifies encoding (charset) of saved CSV files. If it is not set, the UTF-8 charset will be used.

Save Modes - Spark DataFrameWriter also has a method mode() to specify SaveMode. the argument to this method are 
1. overwrite – mode is used to overwrite the existing file
2. append – To add the data to the existing file
3. ignore – Ignores write operation when the file already exists
4. errorifexists or error – This is a default option when the file already exists, it returns an error

```
df5.coalesce(1)\
.write\
.format("csv")\
.option("header", True)\
.option("compression","gzip")\
.mode("Overwrite")\
.save("dbfs:/FileStore/tables/lohi")
```
