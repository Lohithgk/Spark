Spark SQL can automatically infer the schema of a JSON dataset and load it as a Dataset[Row]. This conversion can be done using SparkSession.read.json() - DataFrameReader API.

multiLine is false (default) - Add option to read multi line JSON in the code

```
df = spark\
.read\
.format("json")\
.option("multiline","true")\
.load("dbfs:/FileStore/tables/HumanResources_Employee.json")
```

creates (or replaces if that view name already exists) a lazily evaluated "view" that you can then use like a hive table in Spark SQL. 
createorReplaceTempView is used when you want to store the table for a particular spark session.
Often we might want to store the spark Dataframe as the table and query it, to convert Data frame into temporary view, we use registerTempTable or createOrReplaceTempView (Spark > = 2.0) on our spark Dataframes.

```
df.createOrReplaceTempView("EmployeeData")
```
